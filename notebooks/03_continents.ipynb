```python
iso_codes = [] 
continents = [] 

for iso in africa:
    iso_codes.append(iso) 
    continents.append('Africa')

for iso in asia:
    iso_codes.append(iso) 
    continents.append('Asia')  

for iso in europe:
    iso_codes.append(iso) 
    continents.append('Europe') 

for iso in north_america:
    iso_codes.append(iso) 
    continents.append('North America') 

for iso in south_america:
    iso_codes.append(iso) 
    continents.append('South America') 

for iso in oceania:
    iso_codes.append(iso) 
    continents.append('Oceania')
```

```python
df_continents = pd.DataFrame({
    'ISO' : iso_codes, 
    'Continents': continents
}) 
```
