```python
import pandas as pd 
```

```python
# africa regions and iso codes
north_africa = ['DZ', 'EG', 'LY', 'MA', 'SD', 'TN', 'EH']
west_africa = ['BJ', 'BF', 'CV', 'CI', 'GM', 'GH', 'GN', 'GW', 'LR', 'ML', 'MR', 'NE', 'NG', 'SN', 'SL', 'TG']
central_africa = ['AO', 'CM', 'CF', 'TD', 'CG', 'CD', 'GQ', 'GA', 'ST']
east_africa = ['BI', 'KM', 'DJ', 'ER', 'ET', 'KE', 'MG', 'MW', 'MU', 'RW', 'SC', 'SO', 'SS', 'TZ', 'UG']
southern_africa = ['BW', 'SZ', 'LS', 'MZ', 'NA', 'ZA', 'ZM', 'ZW']
```

```python
iso_codes = []
regions = []

for iso in north_africa:
    iso_codes.append(iso)
    regions.append('North Africa')

for iso in west_africa:
    iso_codes.append(iso)
    regions.append('West Africa')

for iso in central_africa:
    iso_codes.append(iso)
    regions.append('Central Africa')

for iso in east_africa:
    iso_codes.append(iso)
    regions.append('East Africa')

for iso in southern_africa:
    iso_codes.append(iso)
    regions.append('Southern Africa')
```

```python
df_regions = pd.DataFrame({
    'ISO': iso_codes,
    'Region': regions
})

df_regions.to_csv('african_regions.csv', index=False)
print("✓ Saved to 'african_regions.csv'")
```
